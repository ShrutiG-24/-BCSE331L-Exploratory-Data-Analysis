## Data Handling and Cleaning

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Drop duplicates
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

```python
df_clean = df.copy()
```

```python
# Mask for non-empty values
mask = df_clean['INCOME'].notna()

# Remove brackets & parentheses
df_clean.loc[mask, 'INCOME'] = (
    df_clean.loc[mask, 'INCOME']
    .astype(str)
    .str.replace(r'[\[\]\(\)]', '', regex=True)
    .str.strip()
)

# Remove $ signs, commas, and dots
df_clean.loc[mask, 'INCOME'] = (
    df_clean.loc[mask, 'INCOME']
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('.', '', regex=False)
)

# Extract numeric ranges
income_split = df_clean.loc[mask, 'INCOME'].str.extract(r'(\d+)-?(\d+)?')

# Convert to numeric
df_clean['INCOME_MIN'] = pd.to_numeric(income_split[0], errors='coerce')
df_clean['INCOME_MAX'] = pd.to_numeric(income_split[1], errors='coerce')

# Handle case: starts with "-" → min=0
starts_with_dash = df_clean.loc[mask, 'INCOME'].str.startswith('-')
df_clean.loc[starts_with_dash, 'INCOME_MIN'] = 0
df_clean.loc[starts_with_dash, 'INCOME_MAX'] = pd.to_numeric(
    df_clean.loc[starts_with_dash, 'INCOME'].str.replace('-', '', regex=False),
    errors='coerce'
)

# Handle case: ends with "-" → max = column max
ends_with_dash = df_clean.loc[mask, 'INCOME'].str.endswith('-')
col_max = df_clean['INCOME_MAX'].max(skipna=True)
df_clean.loc[ends_with_dash, 'INCOME_MIN'] = pd.to_numeric(
    df_clean.loc[ends_with_dash, 'INCOME'].str.replace('-', '', regex=False),
    errors='coerce'
)
df_clean.loc[ends_with_dash, 'INCOME_MAX'] = col_max

# If max is missing, fill with min
df_clean['INCOME_MAX'] = df_clean['INCOME_MAX'].fillna(df_clean['INCOME_MIN'])

# Average for clean value
df_clean['INCOME_CLEAN'] = df_clean[['INCOME_MIN', 'INCOME_MAX']].mean(axis=1)

# New categorical column
df_clean['Income Range'] = df_clean.apply(
    lambda row: f"{int(row['INCOME_MIN'])}-{int(row['INCOME_MAX'])}"
    if pd.notna(row['INCOME_MIN']) and pd.notna(row['INCOME_MAX'])
    else np.nan,
    axis=1
)
print(df_clean[['INCOME_MIN', 'INCOME_MAX', 'Income Range']]
      .sample(10, random_state=42)
      .to_string(index=False))

```

```python
cols_to_clean = [
    'MARITAL.STATUS', 'EDUCATION', 'OCCUPATION',
    'AREA', 'HOUSEHOLD.SIZE', 'UNDER18',
    'HOUSEHOLDER', 'HOME.TYPE', 'ETHNIC.CLASS', 'LANGUAGE'
]
df_clean[cols_to_clean] = df_clean[cols_to_clean].ffill().bfill()
df_clean.isna().sum()
```

## Univariate Analysis

```python
sorted_ranges = sorted(df_clean['Income Range'].unique(), key=lambda x: int(x.split('-')[0]))

plt.figure(figsize=(12, 8))
sns.countplot(y='Income Range', data=df_clean, order=sorted_ranges, hue='Income Range', palette='coolwarm', legend=False)
plt.title('Frequency of Income Ranges (22BDS0298)')
plt.xlabel('Count')
plt.ylabel('Income Range')
plt.tight_layout()
plt.show()
```

Based on the bar chart, the most common income range in the dataset is 0-10000, which indicates a high concentration of individuals in the lowest income bracket. The least frequent income range is 15000-20000, suggesting that fewer individuals fall into that specific income range.

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df_clean, x='INCOME_CLEAN', fill=True, color='teal')
plt.title('Density Plot of Cleaned Income')
plt.xlabel('Cleaned Income ($)')
plt.ylabel('Density')
plt.tight_layout()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df_clean, x='INCOME_CLEAN', bins=20, hue='Income Range', palette='viridis')
plt.xticks(np.arange(10000, 80000, 5000), rotation=45)

plt.title('Distribution of Cleaned Income(22BDS0298)')
plt.xlabel('Income')
plt.ylabel('Frequency')
```

A traditional histogram is designed for continuous numerical data, where the data is grouped into bins and the height of each bar represents the frequency. INCOME column, however, contains categorical range.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_clean[['INCOME_MIN', 'INCOME_MAX', 'INCOME_CLEAN']], palette=['skyblue', 'lightgreen', 'salmon'])
plt.title('Box Plots for Numerical Income Attributes')
plt.ylabel('Income Value')
plt.xlabel('Attribute')
plt.tight_layout()
```

 All three plots show a similar distribution, with no significant outliers.

```python
categorical_cols = ['SEX', 'MARITAL.STATUS', 'AGE', 'EDUCATION', 'OCCUPATION',
                    'AREA', 'DUAL.INCOMES', 'HOUSEHOLD.SIZE', 'UNDER18', 'HOUSEHOLDER',
                    'HOME.TYPE', 'ETHNIC.CLASS', 'LANGUAGE']

# Assuming df_clean is your DataFrame
for col in categorical_cols:
    print(f"\nBAR CHART for '{col}'")

    counts = df_clean[col].value_counts()


    # Create a bar chart for visualization with count labels
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(y=col, data=df_clean, order=counts.index, palette='coolwarm', hue=col, legend=False)

    # Add count labels to the bars
    for i, v in enumerate(counts):
        ax.text(v + 0.1, i, str(v), color='black', va='center')

    plt.title(f'Frequency of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```
