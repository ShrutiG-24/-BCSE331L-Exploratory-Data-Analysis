## Data Handling and Cleaning

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Drop duplicates
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

```python
df_clean = df.copy()
```

```python
# Mask for non-empty values
mask = df_clean['INCOME'].notna()

# Remove brackets & parentheses
df_clean.loc[mask, 'INCOME'] = (
    df_clean.loc[mask, 'INCOME']
    .astype(str)
    .str.replace(r'[\[\]\(\)]', '', regex=True)
    .str.strip()
)

# Remove $ signs, commas, and dots
df_clean.loc[mask, 'INCOME'] = (
    df_clean.loc[mask, 'INCOME']
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('.', '', regex=False)
)

# Extract numeric ranges
income_split = df_clean.loc[mask, 'INCOME'].str.extract(r'(\d+)-?(\d+)?')

# Convert to numeric
df_clean['INCOME_MIN'] = pd.to_numeric(income_split[0], errors='coerce')
df_clean['INCOME_MAX'] = pd.to_numeric(income_split[1], errors='coerce')

# Handle case: starts with "-" → min=0
starts_with_dash = df_clean.loc[mask, 'INCOME'].str.startswith('-')
df_clean.loc[starts_with_dash, 'INCOME_MIN'] = 0
df_clean.loc[starts_with_dash, 'INCOME_MAX'] = pd.to_numeric(
    df_clean.loc[starts_with_dash, 'INCOME'].str.replace('-', '', regex=False),
    errors='coerce'
)

# Handle case: ends with "-" → max = column max
ends_with_dash = df_clean.loc[mask, 'INCOME'].str.endswith('-')
col_max = df_clean['INCOME_MAX'].max(skipna=True)
df_clean.loc[ends_with_dash, 'INCOME_MIN'] = pd.to_numeric(
    df_clean.loc[ends_with_dash, 'INCOME'].str.replace('-', '', regex=False),
    errors='coerce'
)
df_clean.loc[ends_with_dash, 'INCOME_MAX'] = col_max

# If max is missing, fill with min
df_clean['INCOME_MAX'] = df_clean['INCOME_MAX'].fillna(df_clean['INCOME_MIN'])

# Average for clean value
df_clean['INCOME_CLEAN'] = df_clean[['INCOME_MIN', 'INCOME_MAX']].mean(axis=1)

# New categorical column
df_clean['Income Range'] = df_clean.apply(
    lambda row: f"{int(row['INCOME_MIN'])}-{int(row['INCOME_MAX'])}"
    if pd.notna(row['INCOME_MIN']) and pd.notna(row['INCOME_MAX'])
    else np.nan,
    axis=1
)
print(df_clean[['INCOME_MIN', 'INCOME_MAX', 'Income Range']]
      .sample(10, random_state=42)
      .to_string(index=False))

```

```python
cols_to_clean = [
    'MARITAL.STATUS', 'EDUCATION', 'OCCUPATION',
    'AREA', 'HOUSEHOLD.SIZE', 'UNDER18',
    'HOUSEHOLDER', 'HOME.TYPE', 'ETHNIC.CLASS', 'LANGUAGE'
]
df_clean[cols_to_clean] = df_clean[cols_to_clean].ffill().bfill()
df_clean.isna().sum()
```
